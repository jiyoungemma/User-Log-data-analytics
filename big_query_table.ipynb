```python
from pandas.io import gbq
```

```python
import pandas as pd

sql = """SELECT user_id, product_id, count(*) AS purchase_count,
CASE
WHEN event_type = 'view' THEN 0
WHEN event_type = 'cart' THEN 0
WHEN event_type = 'purchase' THEN 1
END AS event_count
FROM empirical-weft-350709.CP2_ggpark.log_data
WHERE category_code LIKE 'electronics%'
GROUP BY user_id, product_id, event_count
HAVING event_count >= 1
ORDER BY purchase_count DESC"""
project_id = 'empirical-weft-350709'

df = pd.read_gbq(sql, project_id=project_id, dialect='standard')
```

```python
df_elec = df.drop(columns = 'event_count')
```

```python
df_fin = pd.pivot_table(df_elec, index = ['user_id'], columns = ['product_id'], values = 'purchase_count')
```
